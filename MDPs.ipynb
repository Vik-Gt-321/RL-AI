```python
import pandas as pd
import numpy as np
import scipy.stats as st
import math
import random
import matplotlib.pyplot as plt
```

```python
def value_iteration(T, A, S, P, R):
  V = np.zeros((T+1, len(S)))
  policy = np.zeros((len(S), T))
  
  for t in range(T-1, -1, -1):

    for s in range(len(S)):
      max_val = -np.inf
      max_action = -1
      for a in range(len(A)):

        next_value_function = np.dot(P[a][s], V[t+1])
        reward = R(s, A[a])
        current_value = reward + next_value_function

        if current_value > max_val:
          max_val = current_value
          max_action = A[a]
        
      V[t][s] = max_val
      policy[s][t] = max_action

  return policy, V
```

```python
def reward(s, a):
  return s + a*a

T = 10
S = [1, 2, 3]
A = [1, 2]

P_1 = [[0.5, 0.2, 0.3], 
       [0.6, 0.2, 0.2], 
       [0.1, 0.8, 0.1]]
P_2 = [[0.3, 0.3, 0.4], 
       [0.9, 0.05, 0.05], 
       [0.7, 0.2, 0.1]]

P = [P_1, P_2]
P = np.array(P)
```

```python
policy, V = value_iteration(T, A, S, P, reward)
```
