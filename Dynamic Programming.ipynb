```python
import numpy as np
import math
import random
```

```python
g = np.array([[0,1],[1,0]]) #evolution fn
```

```python
r = np.array([[10,5],[2,4]]) #reward
```

```python
rt = np.array([1,2]) #reward terminating
```

```python
sprime=0 #initial state
```

```python
s_pow = 2
a_pow = 2
T=5
```

```python
S = np.arange(0, s_pow)
A = np.arange(0,a_pow)
```

```python
S[0] = sprime
```

```python
V_star = np.zeros((s_pow, T))
```

```python
V_star[:, 0] = rt
```

```python
a_star = np.zeros((a_pow, T))
```

```python
for k in np.arange(1, T):
  for s in S:
    R = []
    for a in A:
      t = r[s, a] + V_star[g[s, a], k-1]
      R.append(t)

    V_star[s, k] = np.max(R)
    a_star[s, T-k] = np.argmax(R)
```
